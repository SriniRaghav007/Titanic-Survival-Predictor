```
data= pd.read_csv('train.csv')
```

```
data=data.drop(['Cabin','Name','Ticket'],axis=1)
```

```
data.isnull().sum()
```

```
data=data.dropna(axis=0)
```

```
data.isnull().sum()
```

```
data.corr()
```

```
data.info()
```

```
data
```

```
x=data.iloc[:,2:].values
y=data.iloc[:,1].values
```

```
x[0]
```

```
y
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,1] = le.fit_transform(x[:, 1])
```

```
x[0]
```

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
x= np.array(ct.fit_transform(x))
```

```

x[0]
```

```

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x)
```

```
X_train[0]
```

```
ann = tf.keras.models.Sequential()
```

```
ann.add(tf.keras.layers.Dense(units=8, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=8, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=8, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy','mse'])
```

```
ann.fit(X_train, y, batch_size = 32, epochs = 100)
```

```
x_test=pd.read_csv('test.csv')
```

```
x_test.isnull().sum()
```

```
x_test=x_test.drop(['Cabin','Name','Ticket'],axis=1)
```

```
x_test=x_test.fillna(x_test.mean())
```

```
x_test.info()
```

```
xt=x_test.iloc[:,1:].values
```

```
xt[:,1] = le.fit_transform(xt[:, 1])
```

```
xt= np.array(ct.transform(xt))
xt = sc.transform(xt)
```

```
y_pred = ann.predict(xt)
```

```
len(y_pred)
```

```
len(x_test['PassengerId'])
```

```
pred=[np.round(i) for i in y_pred]
```
